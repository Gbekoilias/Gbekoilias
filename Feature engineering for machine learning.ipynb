# Working with latitudes and longitudes


```
loan_themes_by_region[['lat','lon','geo']].sample(2)
```

```
def manhattan_distance(lat1, lng1, lat2, lng2):
    a = np.abs(lat2 -lat1)
    b = np.abs(lng1 - lng2)
    return a + b
manhattan_distance(18.660048, 73.813969,18.236421,121.722396	)
```

```
def haversine_array(lat1, lng1, lat2, lng2):
    lat1, lng1, lat2, lng2 = map(np.radians, (lat1, lng1, lat2, lng2))
    AVG_EARTH_RADIUS = 6371  # in km
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = np.sin(lat * 0.5) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(lng * 0.5) ** 2
    h = 2 * AVG_EARTH_RADIUS * np.arcsin(np.sqrt(d))
    return h
```

```
haversine_array(18.660048, 73.813969,18.236421,121.722396	)
```

```
def bearing_array(lat1, lng1, lat2, lng2):
    AVG_EARTH_RADIUS = 6371  # in km
    lng_delta_rad = np.radians(lng2 - lng1)
    lat1, lng1, lat2, lng2 = map(np.radians, (lat1, lng1, lat2, lng2))
    y = np.sin(lng_delta_rad) * np.cos(lat2)
    x = np.cos(lat1) * np.sin(lat2) - np.sin(lat1) * np.cos(lat2) * np.cos(lng_delta_rad)
    return np.degrees(np.arctan2(y, x))
```

```
bearing_array(18.660048, 73.813969,18.236421,121.722396	)
```

```
import featuretools as ft
```

```
kiva_ = peru[["id","funded_amount","loan_amount","sector","repayment_interval","partner_id"]]
```

```
kiva_.info()
```

```
cat_feats = ["sector","repayment_interval"]
```

```
import category_encoders as ce

encoder = ce.OrdinalEncoder(cols=cat_feats)
encoded_data_loan_data = encoder.fit_transform(kiva_)
```

```
encoded_data_loan_data.sample()
```

```
encoded_data_loan_data.reset_index(inplace=True)
```

```
loan_theme_ids.sample()
```

```
loan_theme_ids_ = loan_theme_ids[["Loan Theme Type","id","Partner ID"]]
```

```
cat_feats = ["Loan Theme Type","Partner ID"]
```

```
encoder = ce.OrdinalEncoder(cols=cat_feats)
encoded_data_themes = encoder.fit_transform(loan_theme_ids_)
```

```
encoded_data_themes.sample(2)
```

```
encoded_data_themes = encoded_data_themes.rename(columns={'id': 'loan_id'})
```

```
encoded_data_themes.sample()
```

```
encoded_data_themes.sample()
```

```
es = ft.EntitySet(id="loans")
es = es.entity_from_dataframe(entity_id="loans", dataframe=encoded_data_loan_data, index="id",
           variable_types = { "sector": ft.variable_types.Categorical, 
                             "partner_id": ft.variable_types.Categorical,
                             "funded_amount": ft.variable_types.Numeric,
                             "loan_amount": ft.variable_types.Numeric,
                             "repayment_interval": ft.variable_types.Categorical }) 
```

```
es = es.entity_from_dataframe(entity_id = 'themes',
                              variable_types = {
                             "Loan Theme Type": ft.variable_types.Categorical,
                             "Partner ID": ft.variable_types.Categorical,
                              },
                              dataframe = encoded_data_themes, index = "index")
```

```
# Relationship between loans and themes
r_loan_themes = ft.Relationship(es['loans']['id'], es['themes']['loan_id'])

```

```
es = es.add_relationships([r_loan_themes])
```

```
es
```

```
primitives = ft.list_primitives()
primitives[primitives['type'] == 'aggregation'].head(10)
```

```
primitives[primitives['type'] == 'transform'].head(10)
```
