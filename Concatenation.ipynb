```python
# import relevant libraries
import numpy as np
import pandas as pd
```

```python
# create the train and test sets
train = pd.DataFrame({'SN':['a','b','c','d','e'],
                      'value':[0,1,2,3,4],
                      'croc':[69,70,71,72,73]
                     })
test = pd.DataFrame({'SN':['v','w','x','y','z'],
                      'value':[27,26,24,25,28],
                      'croc':[268,297,294,248,212]
                     })
```
